```python
# https://competitions.codalab.org/competitions/16652
import pandas as pd
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from bs4 import BeautifulSoup
from sklearn import preprocessing
```

```python
base_dir_tr='training/'
base_dir_vl='validation/'
```

```python
dtr=pd.read_csv(base_dir_tr+'data_train.csv',header=None)
col_names=['country','sku_id','title','category_lvl_1','category_lvl_2','category_lvl_3','short_description','price','product_type']
dtr.columns=col_names

#### conciseness_train
xtr=open(base_dir_tr+'conciseness_train.labels')
concise=[]
for l in xtr:
    concise.append(int(l))
    
#### clarity_train
xtr=open(base_dir_tr+'clarity_train.labels')
clarity=[]
for l in xtr:
    clarity.append(int(l))

dtr['conci']=concise
dtr['clari']=clarity
```

```python
dvl=pd.read_csv(base_dir_vl+'data_valid.csv',header=None)
col_names=['country','sku_id','title','category_lvl_1','category_lvl_2','category_lvl_3','short_description','price','product_type']
dvl.columns=col_names
```

```python
# Price in the local currency, e.g., "33.0".  
# When country is my, the price is in Malaysian Ringgit.  
# When country is sg, the price is in Singapore Dollar.  
# When country is ph, the price is in Philippine Peso.
```

```python
dtr_p=dtr[['sku_id','country','price']]
dvl_p=dvl[['sku_id','country','price']]
```

```python
#currency 5 june 2017 21:22
cr={'my':3116.92,'ph':268.93,'sg':9617.17}
```

```python
#to_rupiah
dtr_p2=pd.DataFrame()
for c in dtr_p.country.unique():
    dp=dtr_p[dtr_p.country==c]
    dp['price_rupiah']=dp.price*cr[c]
    dtr_p2=dtr_p2.append(dp)
    
dvl_p2=pd.DataFrame()
for c in dvl_p.country.unique():
    dp=dvl_p[dvl_p.country==c]
    dp['price_rupiah']=dp.price*cr[c]
    dvl_p2=dvl_p2.append(dp)
```

```python
scaler = preprocessing.MinMaxScaler(feature_range=(0,10))
dvl_p2['price_rupiah_st'] = scaler.fit_transform(dvl_p2.price_rupiah)
dtr_p2['price_rupiah_st'] = scaler.fit_transform(dtr_p2.price_rupiah)
```

```python
dtr_p2.sort('price_rupiah')
```

```python
dtr2=pd.read_csv(base_dir_tr+'length_string.csv')
dvl2=pd.read_csv(base_dir_vl+'length_string.csv')
```

```python
dtr3=pd.merge(dtr2,dtr_p2,on='sku_id',how='left')
dvl3=pd.merge(dvl2,dvl_p2,on='sku_id',how='left')
```

```python
dtr3[['sku_id','len_title_des','len_short_des','price_rupiah_st']].to_csv(base_dir_tr+'length_string_2.csv')
dvl3[['sku_id','len_title_des','len_short_des','price_rupiah_st']].to_csv(base_dir_vl+'length_string_2.csv')
```
